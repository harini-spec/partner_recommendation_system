```python
import matplotlib.pyplot as plt
import numpy as np
  
#Euclidean Distance:
Names = ["Elbertina", "Angel", "Netti", "Elfie", "Ronda"]
dist = [1.07, 1.1, 1.16, 1.23, 1.3]
percentile = [94.35, 94.22, 93.9, 93.51, 93.17]
plt.plot(dist, percentile, label = "Euclidean")
# plt.show()

# Minkowski Distance:
dist = [0.85, 0.91, 0.94, 1.08, 1.08]
percentile = [95.54, 95.2, 95.06, 94.3, 94.28]
plt.plot(dist, percentile, label = "Minkowski")
# plt.show()

#Cosine Distance:
dist = [0.11, 0.08, 0.06, 0.16, 0.16]
percentile = [99.44, 99.59, 99.69, 99.15, 99.16]
plt.plot(dist, percentile, label = "Cosine")

#Manhattan Distance:
dist = [2.32, 2.08, 2.39, 2.31, 2.63]
percentile = [87.77, 89.05, 87.28, 87.81, 86.12]
plt.plot(dist, percentile, label = "Manhattan")
# plt.show()

plt.xlabel("Distance")
plt.ylabel("Percentile")
plt.legend()
plt.show()
```
